```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

import plotly
import seaborn as sns
```

```python
df = pd.read_csv('./titanic.csv')
df
```

```python
cantsobrevivio = df[df['Survived'] == 1]['Sex'].value_counts()
cantnosobrevivio = df[df['Survived'] == 0]['Sex'].value_counts()
porcentajesi = (cantsobrevivio / cantsobrevivio.sum()) * 100
porcentajeno = (cantnosobrevivio / cantnosobrevivio.sum()) * 100
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
labels = ['Sobrevivio', 'No Sobrevivio']
colors = ['#66b3ff','#ff9999']
ax[0].pie(porcentajesi, labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')
ax[0].set_title('Porcentaje de supervivencia Hombres')
ax[1].pie(porcentajeno, labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')
ax[1].set_title('Porcentaje de supervivencia Mujeres')

plt.show()
```

```python
total_passengers = df.groupby(['Sex', 'Pclass']).size().unstack()
```

```python
total_passengers.plot(kind='bar', stacked=True)
plt.title('Total de pasajeros Pclass por genero')
plt.xlabel('Genero')
plt.ylabel('Total de pasajeros')
plt.xticks(rotation=0)  # Rotar etiquetas del eje x
plt.legend(title='PClass')
plt.show()
```
