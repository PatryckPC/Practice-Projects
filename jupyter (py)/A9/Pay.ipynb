```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('./titanic.csv')
df
```

```python
pasajeros = df.groupby('Sex').agg(
    {
        'PassengerId':'count'
    }
)
pasajeros
```

```python
pasajeros.plot.pie(
    y='PassengerId',
    startangle=90,
    autopct='%1.2f%%'
    
)
```

```python
fig = px.pie(
    pasajeros,
    names='Sex',
    values='PassengerId'
)
```
