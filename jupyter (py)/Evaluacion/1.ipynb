```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

import plotly
import seaborn as sns
```

```python
df = pd.read_csv('./comparasion_algoritmos.csv')
df
```

```python
plt.figure(figsize=(12, 6))

width = 0.35
x = range(len(df['dataset']))

plt.bar(x, df['algoritmo1'], width, label='Algoritmo 1', color='blue')
plt.bar([p + width for p in x], df['algoritmo2'], width, label='Algoritmo 2', color='orange')

plt.xlabel('Categora')
plt.ylabel('Rendimiento')
plt.title('Evaluacin de Algoritmo1 vs Algoritmo2')

plt.xticks([p + width / 2 for p in x], df['dataset'], rotation=90)

plt.legend()

plt.show()
```
