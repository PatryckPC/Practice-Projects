```python
import pandas as pd
```

```python
df = pd.read_csv('./titanic.csv')
df
```

```python
pasajeros = df.groupby('Sex').count()
pasajeros
```

```python
pasajeros = df.groupby('Sex').agg(
    {
        'PassengerId' : 'count'
    }
)
pasajeros
```

```python
pasajeros.plot.bar(
    y = 'PassengerId',
    width = 0.4,
    title = 'Numero de pasajeros por genero',
    legend = False
)
```
