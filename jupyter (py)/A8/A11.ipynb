```python
import pandas as pd
```

```python
df = pd.read_csv('./titanic.csv')
df
```

```python
pasajeros = df.groupby('Embarked').count()
pasajeros
```

```python
pasajeros = df.groupby('Embarked').agg(
    {
        'PassengerId' : 'count'
    }
)
pasajeros
```

```python
pasajeros.plot.bar(
    y = 'PassengerId',
    width = 0.4,
    title = 'Numero de pasajeros por Embarked',
    legend = False
)
```

```python
import seaborn as sns
```

```python
sns.barplot(
    pasajeros, 
    x='Embarked',
    y='PassengerId',
    width = 0.2
).set(
    title='Numero de pasajeros por Embarked',
)
```

```python
import matplotlib.pyplot as plt
```

```python
pasajeros2 = pasajeros.reset_index()
pasajeros2
```

```python
plt.barh(
    pasajeros2['Embarked'],
    pasajeros2['PassengerId']
)
plt.title('Numero de pasajeros por Embarked')
```
